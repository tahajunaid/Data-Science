```python
import PyPDF2 
import time
import pandas as pd
import tabula
import os

#change directory to where pdf is saved
os.chdir('C:\\Users\\Taha\\Desktop')
os.getcwd()

pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', None)


def findCutoff(category,page_no):
    
    df2 = tabula.read_pdf("college_cutoffs.pdf",pages=[page_no+1])
    rank=""
    for k in range(0,len(df2)):
        if(category=='GOPENH' or category =='GOPENS'):
            if('GOPENH' in df2[k]):
                rank=df2[k]['GOPENH'][0]
                break
            elif ('GOPENS' in df2[k]):
                rank=df2[k]['GOPENS'][0]
                break
        else:
            if(category in df2[k]):
                rank=df2[k][category][0]
                break
    rank=rank.split("\r")
    return rank[0]

def findColleges(category):
    
    pdfFileObj = open('college_cutoffs.pdf', 'rb') 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
    count=0
    df=pd.DataFrame(columns=['clg_id','name','cutoff'])
    
    for i in range(1,pdfReader.numPages):
        print(f'Scanning Page : {i+1}\r', end="")
        pageObj = pdfReader.getPage(i) 
        page=pageObj.extractText()
        res = [i for i in range(len(page)) if page.startswith("Savitribai Phule", i)]


        for j in range(0,len(res)):

            if(res!=[]):
                s_ind=res[j]

                while(1):

                    if page[s_ind:s_ind+6]=="Status":
                        break
                    s_ind=s_ind-1

                branch=page[s_ind-18:s_ind-1] 


                if(branch=="Civil Engineering"):
                    #print("Page No. : ",i+1)
                    course_id=page[s_ind-30:s_ind-21]
                    clg_id=course_id[0:4]
                    #print("College ID : ",clg_id)
                    name_ind=s_ind-31

                    while(1):

                        if page[name_ind:name_ind+4]==clg_id:
                            #print("College Name : ",page[name_ind+7:s_ind-31])
                            break
                        name_ind=name_ind-1

                    rank=findCutoff(category,i)
                    if(rank!=""):
                        df.loc[df.shape[0]]=[clg_id,page[name_ind+7:s_ind-31],rank]
                        count=count+1



    print("Total Records Found : ",count)

    df[['cutoff']] = df[['cutoff']].apply(pd.to_numeric)
    sdf=df.sort_values(['cutoff'])
    sdf.reset_index(inplace=True,drop=True)
    top_10_civil=sdf.head(10)
    display(top_10_civil)
```

```python
category=input("Enter a valid category : ")
findColleges(category)
```
