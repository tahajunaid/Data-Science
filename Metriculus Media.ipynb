```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
sns.set()

%matplotlib inline
plt.style.use('ggplot')
```

```python
def getData(url):

    r = requests.get(url)
    data=r.json()
    data=pd.DataFrame.from_dict(data['states_daily'])
    data['date']=pd.to_datetime(data['date'])
    data.set_index('date', inplace=True)
    data= data.sort_index()
    data=data.apply(pd.to_numeric, errors='ignore')
    return data
```

```python
getData("https://api.covid19india.org/states_daily.json")
```

```python
def Q1_1(json_file_path, start_date, end_date):
    
    data=getData(json_file_path)
    
    start_date=pd.to_datetime(start_date)
    end_date=pd.to_datetime(end_date)
    
    df=data.loc[start_date:end_date].copy()
    
    dfc=df[df['status']=='Confirmed'].copy()
    dfr=df[df['status']=='Recovered'].copy()
    dfd=df[df['status']=='Deceased'].copy()
    
    confirmed_count=dfc['tt'].sum() 
    recovered_count=dfr['tt'].sum() 
    deceased_count=dfd['tt'].sum() 

    print('confirmed_count: ',confirmed_count, 'recovered_count:',recovered_count, 'deceased_count: ',deceased_count)
    return confirmed_count, recovered_count, deceased_count

Q1_1("https://api.covid19india.org/states_daily.json","14-Mar-2020","05-Sept-2020")
```

```python
import datetime as dt
def SlopeIntercept(x_val,y_val):
    x=np.array(x_val).astype(int)
    y=np.array(y_val)
    m=(((np.mean(x)*np.mean(y))-np.mean(x*y))/((np.mean(x))-np.mean(x*x)))
    c=(np.mean(y)-np.mean(x)*m)
    return c,m

def Q3(json_file_path, start_date, end_date):
    
    url=json_file_path
    r = requests.get(url)
    data=r.json()
    data=pd.DataFrame.from_dict(data['states_daily'])
    df=data.copy()

    df['date'] = pd.to_datetime(df['date'])
    df['date']=df['date'].map(dt.datetime.toordinal)

    dfc=df[df['status']=='Confirmed'][['dl','date']].copy()
    dfr=df[df['status']=='Recovered'][['dl','date']].copy()
    dfd=df[df['status']=='Deceased'][['dl','date']].copy()

    confirmed_intercept, confirmed_slope=SlopeIntercept(dfc['dl'],dfc['date'])
    recovered_intercept, recovered_slope=SlopeIntercept(dfr['dl'],dfr['date'])
    deceased_intercept, deceased_slope=SlopeIntercept(dfd['dl'],dfd['date'])
    
    print(' \nConfirmed :')
    print('Intercept: {0} , Slope: {1}'.format(confirmed_intercept, confirmed_slope))
    print('\nRecovered :')
    print('Intercept: {0} , Slope: {1}'.format(recovered_intercept, recovered_slope))
    print('\nDeceased :')
    print('Intercept: {0} , Slope: {1}'.format(deceased_intercept,deceased_slope))
    return confirmed_intercept, confirmed_slope,recovered_intercept,recovered_slope, deceased_intercept, deceased_slope
    

Q3("https://api.covid19india.org/states_daily.json","14-Mar-2020","05-Sept-2020") 
```
